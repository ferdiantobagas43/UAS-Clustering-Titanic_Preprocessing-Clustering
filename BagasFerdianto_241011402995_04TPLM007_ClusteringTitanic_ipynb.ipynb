### Tujuan
Proyek ini bertujuan melakukan preprocessing terhadap dataset Titanic dan mengimplementasikan algoritma K-Means Clustering untuk mengelompokkan data berdasarkan karakteristik yang dimiliki setiap penumpang.

## 1. Import Library

Library yang digunakan pada proyek ini meliputi pandas, numpy, matplotlib, seaborn, dan scikit-learn untuk proses preprocessing, clustering, evaluasi, serta visualisasi data.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

## 2. Memuat Dataset

Dataset Titanic dimuat menggunakan library pandas. Dataset ini berisi informasi mengenai penumpang Titanic seperti usia, kelas penumpang, jenis kelamin, tarif, dan status keselamatan.

```python
df = pd.read_csv("titanic.csv")

df.head()
```

## 3. Eksplorasi Dataset

Tahap ini bertujuan untuk mengetahui struktur dataset, tipe data, jumlah data, serta mendeteksi adanya missing value sebelum dilakukan preprocessing.

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["Cabin"])
```

## 4. Cleaning Data

Cleaning data dilakukan dengan menghapus kolom yang memiliki banyak missing value, mengisi nilai yang kosong menggunakan median dan modus, serta memeriksa apakah terdapat data duplikat.

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop(columns=[
    "PassengerId",
    "Name",
    "Ticket"
])
```

```python
df.columns.tolist()
```

## 5. Encoding Data

Kolom kategorikal seperti Sex dan Embarked diubah menjadi data numerik menggunakan One Hot Encoding agar dapat diproses oleh algoritma K-Means.

```python
df.head()
```

## 6. Standardisasi Data

Standardisasi dilakukan menggunakan StandardScaler agar seluruh fitur memiliki skala yang sama sehingga proses clustering menjadi lebih optimal.

```python
scaler = StandardScaler()

scaled = scaler.fit_transform(df)
```

## 7. Menentukan Jumlah Cluster (Elbow Method)

Elbow Method digunakan untuk menentukan jumlah cluster yang optimal berdasarkan nilai WCSS (Within Cluster Sum of Squares).

```python
wcss = []

for i in range(1,11):
    model = KMeans(
        n_clusters=i,
        random_state=42
    )

    model.fit(scaled)

    wcss.append(model.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11),wcss,marker="o")
plt.xlabel("Jumlah Cluster")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

## 8. Implementasi K-Means

Algoritma K-Means diterapkan menggunakan jumlah cluster yang telah ditentukan dari hasil evaluasi.

```python
kmeans = KMeans(
    n_clusters=6,
    random_state=42
)

cluster = kmeans.fit_predict(scaled)

df["Cluster"] = cluster
```

```python
df.head()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(scaled, cluster)

print("Silhouette Score:", score)
```

## 9. Evaluasi (Silhouette Score)

Silhouette Score digunakan untuk mengukur kualitas hasil clustering. Semakin tinggi nilainya, semakin baik pemisahan antar cluster.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

for k in range(2, 8):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled)
    score = silhouette_score(scaled, labels)
    print(f"k = {k}, Silhouette Score = {score:.4f}")
```

```python
kmeans = KMeans(n_clusters=7, random_state=42)

cluster = kmeans.fit_predict(scaled)

df["Cluster"] = cluster
```

```python
score = silhouette_score(scaled, cluster)
print(score)
```

## 10. Visualisasi Hasil Clustering

Hasil clustering divisualisasikan menggunakan PCA (Principal Component Analysis) sehingga data berdimensi banyak dapat ditampilkan dalam bentuk dua dimensi.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
hasil = pca.fit_transform(scaled)

plt.figure(figsize=(8,6))
plt.scatter(hasil[:,0], hasil[:,1], c=cluster, cmap='viridis')
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Hasil Clustering K-Means")
plt.show()
```

## 11. Analisis Cluster

Tahap ini bertujuan untuk mengetahui karakteristik setiap cluster berdasarkan nilai rata-rata masing-masing fitur.

```python
df.groupby("Cluster").mean()
```

## 12. Kesimpulan

Berdasarkan hasil preprocessing dan penerapan algoritma K-Means, dataset Titanic berhasil dikelompokkan menjadi 7 cluster. Tahapan preprocessing seperti penanganan missing value, encoding data kategorikal, dan standardisasi berhasil meningkatkan kualitas data sebelum proses clustering. Berdasarkan evaluasi menggunakan Silhouette Score, nilai terbaik diperoleh pada k = 7 dengan nilai sebesar 0,3001.
